```python
import pandas as pd 
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt
import seaborn as sns
import scipy as sp
import json
import os
```

```python
import glob
import json

files = glob.glob('data/*.json')

combined_data = []
for filename in files:
    with open(filename, 'r', encoding='utf-8') as file:
        data = json.load(file)
        combined_data.append(data)

with open('result.json', 'w', encoding='utf-8') as outfile:
    json.dump(combined_data, outfile, indent=4)
```

```python
# read the data
data = json.load(open('result.json', encoding='utf8'))

df = pd.DataFrame(data)
df.info()
```

```python
mask = df.isna().sum() / len(df)  < 0.4
mask
```

```python
reduced_df = df.loc[:, mask]
reduced_df.drop(['Museum number', 'Bibliographic references', 'Object Type'], axis=1, inplace=True)
reduced_df.tail()
```

```python
reduced_df['Description_ex'] = reduced_df['Description'].str[0]
reduced_df['Materials_ex'] = reduced_df['Materials'].str[0]
# should I remove Object Type since it is redundant with artifact name?
reduced_df['Findspot_ex'] = reduced_df['Findspot'].str[0]
reduced_df['Dimensions_1'] = reduced_df['Dimensions'].str[0]
reduced_df['Dimensions_2'] = reduced_df['Dimensions'].str[1]
reduced_df['Location_ex'] = reduced_df['Location'].str[0]
reduced_df['Acquisition_name_ex'] = reduced_df['Acquisition name'].str[0]
reduced_df['Acquisition_date_ex'] = reduced_df['Acquisition date'].str[0]
reduced_df['Department_ex'] = reduced_df['Department'].str[0]
reduced_df['Cultures/periods_ex'] = reduced_df['Cultures/periods'].str[0]

to_drop = ['Description', 'Materials', 'Findspot', 'Dimensions', 'Location', 'Acquisition name', 'Acquisition date', 'Department', 'Registration number', 'Cultures/periods']

```

```python
reduced_df.drop(to_drop, axis=1, inplace=True)
reduced_df.head()
```

```python
# Define the source folder that contains the images
source_folder = "photo"

# Get the list of image filenames
image_filenames = os.listdir(source_folder)

# Create a new column in the dataframe with the image paths
reduced_df["image_path"] = reduced_df.apply(lambda row: os.path.join(source_folder, image_filenames[row.name]), axis=1)

# Print the first five rows of the dataframe
reduced_df.tail()
```
